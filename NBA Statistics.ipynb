```python
filepath = './games_details.csv'
df_games_details = pd.read_csv(filepath)
filepath = './games.csv'
df_games_unfiltered = pd.read_csv(filepath)
df_games = df_games_unfiltered[(df_games_unfiltered['GAME_DATE_EST'] <= '2019-07-01') | (df_games_unfiltered['GAME_DATE_EST'] >= '2020-07-01')]



games = []

def count_up_games(game_id, home_id, visitor_id, home_winner, i):
    total_home = df_games_details[(df_games_details.GAME_ID == game_id) & (df_games_details.TEAM_ID == home_id)]
    total_away = df_games_details[(df_games_details.GAME_ID == game_id) & (df_games_details.TEAM_ID == visitor_id)]
    total_bigs_home = total_home.loc[(total_home["START_POSITION"] == 'F') | (total_home["START_POSITION"] == 'C'), "FG3M"].sum() 
    total_bigs_away = total_away.loc[(total_away["START_POSITION"] == 'F') | (total_away["START_POSITION"] == 'C'), "FG3M"].sum()
    total_smalls_home = total_home.loc[(total_home["START_POSITION"] == 'G'), "FG3M"].sum() 
    total_smalls_away = total_away.loc[(total_away["START_POSITION"] == 'G'), "FG3M"].sum() 
    
    if (total_home["FG3M"].sum() > total_away["FG3M"].sum()) & (home_winner == 1):
        games[i].append(1) ##puts a 1 in the array where the winning team has more 3-pointers
    elif ((total_home["FG3M"].sum() < total_away["FG3M"].sum()) & (home_winner == 0)):
        games[i].append(1)
    else:
        games[i].append(0) ##games[i][5]
##now we do offensive rebounds
    if (total_home["OREB"].sum() > total_away["OREB"].sum()) & (home_winner == 1):
        games[i].append(1) ##puts a 1 in the array where the winning team has more 3-pointers
    elif ((total_home["OREB"].sum() < total_away["OREB"].sum()) & (home_winner == 0)):
        games[i].append(1) ##for offensive rebounds, same as 3pointers
    else:
        games[i].append(0) ##games[i][6]
    if (total_home["DREB"].sum() > total_away["DREB"].sum()) & (home_winner == 1):
        games[i].append(1) ##puts a 1 in the array where the winning team has more 3-pointers
    elif ((total_home["DREB"].sum() < total_away["DREB"].sum()) & (home_winner == 0)):
        games[i].append(1) ##for defensive rebounds, same as 3pointers
    else:
        games[i].append(0) ##games[i][7]
    if (total_home["TO"].sum() < total_away["TO"].sum()) & (home_winner == 1):
        games[i].append(1) ##puts a 1 in the array where the winning team has more 3-pointers
    elif ((total_home["TO"].sum() > total_away["TO"].sum()) & (home_winner == 0)):
        games[i].append(1) ##for to rebounds, opposite as 3pointers, since we don't want them
    else:
        games[i].append(0) ##games[i][8]
    if ((total_bigs_home > total_bigs_away) & (home_winner == 1)):
        games[i].append(1) ##puts a 1 in the array where the winning team has more 3-pointers
    elif ((total_bigs_home < total_bigs_away) & (home_winner == 0)):
        games[i].append(1) ##for offensive rebounds, same as 3pointers
    else:
        games[i].append(0) ##games[i][9]
    if ((total_smalls_home > total_smalls_away) & (home_winner == 1)):
        games[i].append(1) ##puts a 1 in the array where the winning team has more 3-pointers
    elif ((total_smalls_home < total_smalls_away) & (home_winner == 0)):
        games[i].append(1) ##for offensive rebounds, same as 3pointers
    else:
        games[i].append(0) ##games[i][10]
    
    


for index, row in df_games.iterrows():
    games.append([row["GAME_ID"],row["HOME_TEAM_ID"],row["VISITOR_TEAM_ID"],row["HOME_TEAM_WINS"], row["SEASON"]])
for i in range(len(games)):
     count_up_games(games[i][0],games[i][1],games[i][2],games[i][3], i)
     

##ran this once, we now have basically all of our rows, took a couple minutes to run everything
```
